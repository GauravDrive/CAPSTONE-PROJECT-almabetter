##**Problem no 5: What is the Casualities rate of terrorism**

````python
gtd.head()
````

````python
Number_of_people_killed = gtd[['Attack_Type', 'nkill']].groupby(["Attack_Type"],axis=0).sum()
Number_of_people_wounded = gtd[['Attack_Type','nwound']].groupby(["Attack_Type"],axis=0).sum()

# Create a figure with two subplots in one row
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))

# Plot the number of people killed on the first subplot
sns.barplot(x=Number_of_people_killed.index, y=Number_of_people_killed.nkill.values, palette="cubehelix", ax=ax1)
ax1.set_title('Number of People Killed in each attack type')
ax1.set_xlabel('Types of attacks')
ax1.set_ylabel('Number of people killed')
ax1.tick_params(axis='x', labelrotation=90)
for i in range(len(Number_of_people_killed)):
    ax1.annotate(Number_of_people_killed.values[i], 
                 (i, Number_of_people_killed.values[i]), 
                 ha='center', va='bottom')

# Plot the number of people wounded on the second subplot
sns.barplot(x=Number_of_people_wounded.index, y=Number_of_people_wounded.nwound.values, palette="cubehelix", ax=ax2)
ax2.set_title('Number of People Wounded in each attack type')
ax2.set_xlabel('Types of attacks')
ax2.set_ylabel('Number of people wounded')
ax2.tick_params(axis='x', labelrotation=90)
for i in range(len(Number_of_people_wounded)):
    ax2.annotate(Number_of_people_wounded.values[i], 
                 (i, Number_of_people_wounded.values [i]), 
                 ha='center', va='bottom')

# Display the chart
plt.show()
````
