#**Problem no 7: what is the rate of terrorism increase during past years**

```python
corr_gtd = Global_Impact.groupby('Year')['nkill'].sum().reset_index()
years = corr_gtd['Year'].tolist()
nkill = corr_gtd['nkill'].tolist()

# Create a color map based on the values
import matplotlib.cm as cm
cmap = cm.get_cmap('viridis')
normalized_nkill = (nkill - np.min(nkill)) / (np.max(nkill) - np.min(nkill))
colors = cmap(normalized_nkill)

# Plot the bar chart
fig, ax = plt.subplots()
ax.bar(years, nkill, color=colors)

# Set the title and labels
ax.set_title('Total Number of Kills per Year')
ax.set_xlabel('Year')
ax.set_ylabel('Number of Kills')

plt.show()
```

```python
plt.fill_between(years, nkill, color='red', alpha=0.5)
plt.title('Total Number of Kills per Year')
plt.xlabel('Year')
plt.ylabel('Number of Kills')
plt.show()
```
